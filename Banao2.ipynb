```python
# Installing required libraries

from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import Select
import time
```

```python
driver=webdriver.Chrome()  # setting up the driver
```

```python
df=pd.read_csv('twitter_links.csv') # reading the file.
df
```

```python
bio=[]
following=[]
followers=[]
location=[]
site=[]
for i in range(len(df)):
    url=df.iloc[i][0] 
    driver.get(url) # getting the link 
    time.sleep(5) # to load the page completely
    
    # using try-catch block so that if data is unavailable code keep on executing.
    try:
        bio.append(driver.find_element(By.XPATH,"//div[@data-testid='UserDescription']//span").text) 
    except:
        bio.append('')
    try:
        following.append(driver.find_elements(By.XPATH,"//div[@class='css-1dbjc4n r-13awgt0 r-18u37iz r-1w6e6rj']//span[@class='css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0']")[0].text)
    except:
        following.append('')
    try:
        followers.append(driver.find_elements(By.XPATH,"//div[@class='css-1dbjc4n r-13awgt0 r-18u37iz r-1w6e6rj']//span[@class='css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0']")[2].text)
    except:
        followers.append('')
    try:
        location.append(driver.find_element(By.XPATH,"//span[@data-testid='UserLocation']").text)
    except:
        location.append('')
    try:
        site.append(driver.find_element(By.XPATH,"//div[@data-testid='UserProfileHeader_Items']//a").text)
    except:
        site.append('')
```

```python
df1=pd.DataFrame(list(zip(bio,following,followers,location,site)),columns=['Bio','Following Count','Followers Count','Location','Website'])
df1
```

```python
df1.to_csv('Twitter_Bio.csv')
```
